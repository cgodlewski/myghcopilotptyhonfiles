```python
import yfinance as yf
import pandas as pd

# Load ticker symbols from an Excel file
ticker_df = pd.read_excel(r"C:\Users\GODLEWSKI\OneDrive - unistra.fr\DATA\THOMSON REUTERS\NEW 2022 + 2024\europe tickers for yfinance.xlsx")  # Replace with your file path
tickers = ticker_df['Ticker Symbol'].tolist()[:100]  # Get the first 100 tickers

# Define the start and end dates for the historical data
start_date = '1999-01-01'
end_date = '2024-01-01'

# Initialize a dictionary to store the adjusted close prices
adjusted_close_prices = {}

# Fetch historical data for each ticker
for ticker in tickers:
    try:
        # Fetch data for the ticker within the date range
        data = yf.download(ticker, start=start_date, end=end_date)
        # Store the adjusted close prices
        adjusted_close_prices[ticker] = data['Adj Close']
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")

# Calculate daily returns for each ticker
daily_returns = {ticker: prices.pct_change().dropna() for ticker, prices in adjusted_close_prices.items()}

# Convert the daily returns dictionary to a DataFrame
daily_returns_df = pd.DataFrame(daily_returns)

# Save the daily returns to a new Excel file
daily_returns_df.to_excel('daily_returns_1999_2024.xlsx')
```

# Script 1: Validate Tickers

This script takes a list of ticker symbols from an Excel file and checks each one against Yahoo Finance to determine if itâ€™s valid. 
It outputs two lists: one for valid tickers and another for invalid tickers, each saved to separate Excel files.

```python
import yfinance as yf
import pandas as pd
import time  

def validate_tickers(ticker_list):
    valid_tickers = []
    invalid_tickers = []

    for ticker in ticker_list:
        try:
            info = yf.Ticker(ticker).info
            if info:
                valid_tickers.append(ticker)
        except Exception as e:
            invalid_tickers.append(ticker)

    return valid_tickers, invalid_tickers

if __name__ == "__main__":
    # Load ticker symbols from an Excel file
    ticker_df = pd.read_excel(r"C:\Users\GODLEWSKI\OneDrive - unistra.fr\DATA\THOMSON REUTERS\NEW 2022 + 2024\europe tickers for yfinance.xlsx")
    tickers = ticker_df['Ticker Symbol'].tolist()

    # Validate tickers
    valid, invalid = validate_tickers(tickers)

    for ticker in ticker_list:
    try:
        info = yf.Ticker(ticker).info
        if info:
            valid_tickers.append(ticker)
    except Exception as e:
        invalid_tickers.append(ticker)
    time.sleep(0.1)  # Sleep for 100 milliseconds

    # Save the valid and invalid tickers to separate files
    pd.Series(valid).to_excel('valid_tickers.xlsx', index=False)
    pd.Series(invalid).to_excel('invalid_tickers.xlsx', index=False)
```

# Script 2: Extract Adjusted Returns

This script processes a list of validated ticker symbols in chunks. For each chunk, it fetches historical adjusted close prices from Yahoo Finance, calculates daily returns, and saves the data to an Excel file. 
The script handles the tickers in batches to efficiently manage large numbers of tickers and to avoid hitting rate limits.

```python
import yfinance as yf
import pandas as pd

# Function to fetch adjusted returns for a chunk of tickers
def fetch_adjusted_returns(ticker_list, start_date, end_date):
    adjusted_close_prices = {}
    for ticker in ticker_list:
        try:
            data = yf.download(ticker, start=start_date, end=end_date)
            adjusted_close_prices[ticker] = data['Adj Close']
        except Exception as e:
            print(f"Error fetching data for {ticker}: {e}")
    return adjusted_close_prices

# Load valid tickers from the validation script's output
valid_tickers_df = pd.read_excel('valid_tickers.xlsx')
valid_tickers = valid_tickers_df['Ticker Symbol'].tolist()

# Define the start and end dates for the historical data
start_date = '1999-01-01'
end_date = '2024-01-01'

# Define the size of each chunk
chunk_size = 100

# Calculate the number of chunks needed
num_chunks = len(valid_tickers) // chunk_size + (1 if len(valid_tickers) % chunk_size else 0)

# Loop through the list of tickers in chunks
for i in range(num_chunks):
    start_index = i * chunk_size
    end_index = start_index + chunk_size
    ticker_chunk = valid_tickers[start_index:end_index]

    # Fetch adjusted returns for the current chunk
    adjusted_returns = fetch_adjusted_returns(ticker_chunk, start_date, end_date)

    # Convert the adjusted returns to a DataFrame
    returns_df = pd.DataFrame(adjusted_returns)

    # Save the returns to a new Excel file for each chunk
    returns_df.to_excel(f'daily_returns_chunk_{i+1}.xlsx')

print("All data has been processed and saved.")
```

```python
import yfinance as yf

# Replace 'AAPL' with your desired stock ticker
ticker = yf.Ticker('AAPL')
hist = ticker.history(period='20y')  # Fetches historical data

import matplotlib.pyplot as plt

# Plot the closing prices
plt.figure(figsize=(14, 7))
plt.plot(hist.index, hist['Close'], label='Adjusted Close Price')
plt.title('Historical Adjusted Close Price')
plt.xlabel('Date')
plt.ylabel('Price in USD')
plt.legend()
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Replace 'AAPL' with your desired stock ticker
ticker = yf.Ticker('AAPL')
hist = ticker.history(period='1y')  # Fetches one year of historical data

# Create a figure and a single subplot
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot the adjusted close price on the primary y-axis (left)
ax1.plot(hist.index, hist['Close'], label='Adjusted Close Price', color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Adjusted Close Price', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.legend(loc='upper left')

# Instantiate a second y-axis (right) sharing the same x-axis
ax2 = ax1.twinx()

# Plot the volume as bars on the secondary y-axis (right)
ax2.bar(hist.index, hist['Volume'], label='Volume', color='gray', alpha=0.3)
ax2.set_ylabel('Volume', color='gray')
ax2.tick_params(axis='y', labelcolor='gray')
ax2.legend(loc='upper right')

# Title of the plot
plt.title('Adjusted Close Price and Volume for AAPL')

# Show the plot
plt.show()
```
