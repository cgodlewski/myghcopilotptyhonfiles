```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a new graph
G = nx.Graph()

# List of studies and topics
studies = [
    "Akinci & Olmstead-Rumsey (2018)", "Apergis et al. (2022)", "Behncke (2023)", "Calem et al. (2020)",
    "Cerutti et al. (2017)", "Claessens et al. (2013)", "De Schryder & Opitz (2021)", "Gambacorta & Murcia (2020)",
    "Jiménez et al. (2017)", "Altunbas et al. (2018)", "Davis et al. (2022)", "Ely et al. (2021)", "Gonzalez (2005)",
    "Meuleman & Vander Vennet (2020)", "Olszak et al. (2018)", "Ćehajić & Košak (2021)", "Chen et al. (2022)",
    "Gaganis et al. (2021)"
]

topics = ["Credit Cycles", "Bank Risks", "Bank Efficiency & Profitability"]

# Add nodes for topics with special attributes
topic_colors = ['red', 'green', 'blue']
for topic, color in zip(topics, topic_colors):
    G.add_node(topic, color=color, size=1000)


# Add edges between studies and their topics
edges = [
    ("Akinci & Olmstead-Rumsey (2018)", "Credit Cycles"),
    ("Apergis et al. (2022)", "Credit Cycles"),
    ("Behncke (2023)", "Credit Cycles"),
    ("Calem et al. (2020)", "Credit Cycles"),
    ("Cerutti et al. (2017)", "Credit Cycles"),
    ("Claessens et al. (2013)", "Credit Cycles"),
    ("De Schryder & Opitz (2021)", "Credit Cycles"),
    ("Gambacorta & Murcia (2020)", "Credit Cycles"),
    ("Jiménez et al. (2017)", "Credit Cycles"),
    ("Altunbas et al. (2018)", "Bank Risks"),
    ("Davis et al. (2022)", "Bank Risks"),
    ("Ely et al. (2021)", "Bank Risks"),
    ("Gonzalez (2005)", "Bank Risks"),
    ("Meuleman & Vander Vennet (2020)", "Bank Risks"),
    ("Olszak et al. (2018)", "Bank Risks"),
    ("Ćehajić & Košak (2021)", "Bank Efficiency & Profitability"),
    ("Chen et al. (2022)", "Bank Efficiency & Profitability"),
    ("Davis et al. (2022)", "Bank Efficiency & Profitability"),
    ("Gaganis et al. (2021)", "Bank Efficiency & Profitability")
]

# Add study nodes to the graph with attributes
for study in studies:
    G.add_node(study, color='skyblue', size=500)

# Add edges between studies and their topics
G.add_edges_from(edges)

# Extract position for the nodes
pos = nx.spring_layout(G, k=0.5)

# Increase figure size for better visibility on slides
plt.figure(figsize=(13, 10))

# Draw the nodes
node_colors = [G.nodes[node]['color'] for node in G.nodes()]
node_sizes = [G.nodes[node]['size'] for node in G.nodes()]
nx.draw_networkx_nodes(G, pos, node_color='none', edgecolors=node_colors, node_size=node_sizes, linewidths=2, alpha=0.8)

# Draw the edges
nx.draw_networkx_edges(G, pos, alpha=0.5)

# Draw the labels with increased font size
# Draw the labels with increased font size
for node, (x, y) in pos.items():
    if node in topics:
        plt.text(x, y, node, fontsize=12, fontweight='bold', ha='center')
    else:
        plt.text(x, y, node, fontsize=10, ha='center')

plt.title("Network Visualization of Macroprudential Policy Literature")
plt.show()
```
