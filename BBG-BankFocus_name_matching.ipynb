```python
# define details for matching

from name_matching.name_matcher import NameMatcher

matcher = NameMatcher(top_n=10,
    lowercase=True,
    punctuations=True,
    remove_ascii=True,
    legal_suffixes=False,
    common_words=False,
    preprocess_split=False,
    verbose=True)
```

iterative_sub_string: This metric calculates the similarity between two strings by iteratively finding and comparing substrings.
pearson_ii: This is a variant of the Pearson correlation coefficient, a measure of the linear correlation between two variables. In this case, it’s used to measure the correlation between the character sequences in two names.
bag: This metric treats the names as “bags” of characters and calculates the similarity based on the number of characters they have in common.
fuzzy_wuzzy_partial_string: This metric uses the FuzzyWuzzy library to calculate a similarity score based on partial string matching.
editex: This is a phonetic algorithm that calculates the similarity between names based on how they sound rather than how they are spelled.

If your names have many typographical errors, metrics like discounted_levenshtein, tichy, or typo might be useful. These metrics are based on the number of insertions, deletions, and substitutions needed to transform one string into another.
If your names have variations in word order, you might find fuzzy_wuzzy_token_sort or fuzzy_wuzzy_token_set useful. These metrics tokenize the strings and sort or setify the tokens before comparing.
If your names might have extra or missing words, consider using bag or overlap. These metrics treat the strings as bags of words and compare the contents of the bags.
If your names might be phonetically similar but spelled differently, editex, refined_soundex, or double_metaphone could be helpful. These metrics are based on how the strings sound rather than how they are spelled.
If your names are long and might contain matching substrings, iterative_sub_string could be a good choice. This metric finds and compares substrings within the strings.

```python
# define metrics to match
matcher.set_distance_metrics(['iterative_sub_string', 'discounted_levenshtein', 'bag', 'fuzzy_wuzzy_partial_string', 'editex', 'refined_soundex', 'typo'])
```

```python
# load the data to which the names should be matched
matcher.load_and_process_master_data(column='Matched company name',
                                     df_matching_data=bf_names,
                                     transform=True)
```

```python
# perform the name matching on the data you want matched
matches = matcher.match_names(to_be_matched=bbg_names,
                              column_matching='Company name')
```

```python
# combine the datasets based on the matches
combined = pd.merge(bf_names, matches, how='left', left_index=True, right_on='match_index')
combined = pd.merge(combined, bbg_names, how='left', left_index=True, right_index=True)
```

```python
# Save the combined DataFrame to a CSV file
combined.to_csv('bbg-bf_combined.csv', index=False)
```
