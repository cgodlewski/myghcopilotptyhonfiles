# EXTRACT ALL IDENTIFIERS FROM 1st COLUMN FROM SEVERAL EXCEL FILES (EIKON GRID EXPORT LOANS)

```python
import pandas as pd
import glob

# Path to the Excel files
file_path = "C:\\Users\\GODLEWSKI\\OneDrive - unistra.fr\\DATA\\THOMSON REUTERS\\NEW 2022 + 2024\\Loans\\eikon grid exports loans Europe financial data\\financial variables for grid export *"

# List of all Excel files matching the pattern
files = glob.glob(file_path)

# Initialize an empty list to store the identifiers
identifiers = []

# Loop through each file and extract the first column
for file in files:
    data = pd.read_excel(file, usecols=[0])
    identifiers.append(data)

# Concatenate all the identifiers into a single DataFrame
identifiers_df = pd.concat(identifiers, ignore_index=True)

# Save the identifiers to a CSV file
identifiers_df.to_csv("identifiers.csv", index=False)
```

```python
### chatgpt version of codebook for a file

import pandas as pd

def generate_codebook(file_path, variables=None):
    # Read the file
    df = pd.read_excel(file_path)

    # Basic information
    num_lines = len(df)
    num_columns = len(df.columns)
    column_names = df.columns.tolist()
    data_types = df.dtypes

    # Missing values
    missing_values_percentage = (df.isnull().mean() * 100).round(2)

    # Summary statistics for numerical columns
    summary_statistics = df.describe().transpose()

    # Value counts for categorical variables
    value_counts = {}
    for variable in variables:
        if variable in column_names:
            if df[variable].dtype == 'object':
                value_counts[variable] = df[variable].value_counts().to_dict()
        else:
            print(f"Warning: '{variable}' not found in column names.")

    # Memory usage
    memory_usage = df.memory_usage(deep=True).sum() / (1024 * 1024)  # in MB

    codebook = {
        'Number of lines': num_lines,
        'Number of columns': num_columns,
        'Column names': column_names,
        'Data types': data_types,
        'Missing values (%)': missing_values_percentage,
        'Summary statistics': summary_statistics,
        'Value counts': value_counts,
        'Memory usage (MB)': memory_usage
    }

    return codebook

# Example usage:
file_path = 'your_file.xlsx'  # Replace 'your_file.xlsx' with the actual file path
variables_to_check = ['Variable1', 'Variable2']  # Specify the variables to check for value counts
codebook = generate_codebook(file_path, variables=variables_to_check)
for key, value in codebook.items():
    print(f"{key}: {value}")
```

```python
### mistral version of the codebook

import pandas as pd

def generate_codebook(file_path, variables=None, n_rows=5):
    # Read the file
    df = pd.read_excel(file_path, engine='openpyxl')

    # Basic information
    num_lines = len(df)
    num_columns = len(df.columns)
    column_names = df.columns.tolist()
    data_types = df.dtypes.to_dict()

    # Missing values
    missing_values_percentage = (df.isnull().mean() * 100).round(2).to_dict()

    # Unique values for categorical variables
    categorical_cols = df.select_dtypes(include=['object']).columns
    unique_values = {}
    for col in categorical_cols:
        if variables is not None and col not in variables:
            continue
        unique_values[col] = df[col].nunique()

    # Summary statistics for numerical columns
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    summary_statistics = df[numerical_cols].describe().transpose().to_dict()

    # Correlations for numerical columns
    correlation_matrix = df[numerical_cols].corr()

    # Duplicates
    num_duplicates = df.duplicated().sum()
    duplicate_percentage = (num_duplicates / num_lines) * 100

    # First and last N rows
    first_n_rows = df.head(n_rows)
    last_n_rows = df.tail(n_rows)

    # Memory usage
    memory_usage = df.memory_usage(deep=True).sum() / (1024 * 1024)  # in MB

    codebook = {
        'Number of lines': num_lines,
        'Number of columns': num_columns,
        'Column names': column_names,
        'Data types': data_types,
        'Missing values (%)': missing_values_percentage,
        'Unique values': unique_values,
        'Summary statistics': summary_statistics,
        'Correlations': correlation_matrix,
        'Duplicates': num_duplicates,
        'Duplicate percentage (%)': duplicate_percentage,
        'First N rows': first_n_rows,
        'Last N rows': last_n_rows,
        'Memory usage (MB)': memory_usage
    }

    return codebook

# Example usage:
file_path = 'your_file.xlsx'  # Replace 'your_file.xlsx' with the actual file path
variables_to_check = ['Variable1', 'Variable2']  # Specify the variables to check for value counts, put None without [] to check all
codebook = generate_codebook(file_path, variables=variables_to_check, n_rows=10)
for key, value in codebook.items():
    print(f"{key}: {value}")
```
