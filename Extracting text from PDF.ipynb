```python
from pdfminer.high_level import extract_text

def extract_text_from_pdf(pdf_path):
    # Extract text from the PDF
    text = extract_text(pdf_path)
    return text

# Call the function and print the results
# Note: Replace 'your_document.pdf' with the actual PDF file name
text = extract_text_from_pdf(r"C:\Users\GODLEWSKI\Downloads\Programme Detailed Export (3).pdf")
print("Extracted text:", text)
```

```python
def extract_names_from_text(text):
    # Define a regex pattern to match names after "Authors:" and before "Presenter:", "Discussant:", or a blank line
    pattern = r"Authors:\s*(.*?)(?=Presenter:|Discussant:|\n\n)"

    # Find all matches of the pattern
    matches = re.findall(pattern, text, re.DOTALL)

    # Process matches to remove affiliations and split names
    names = []
    for match in matches:
        # Remove text within parentheses and any trailing spaces
        clean_match = re.sub(r"\s*\([^)]*\)", "", match).strip()
        # Replace newline characters with a space
        clean_match = clean_match.replace('\n', ' ')
        # Split names by comma and strip whitespace
        individual_names = [name.strip() for name in clean_match.split(",")]
        # Extend the names list with individual names
        names.extend(individual_names)

    # Post-process names to remove unwanted text
    names = [re.sub(r'Discussant:.*', '', name) for name in names]
    names = [name.split('\n')[0] for name in names]

    # Convert the list of names to a set to remove duplicates, then convert it back to a list
    names = list(set(names))

    # Return the list of names
    return names
```

```python
def save_names_to_csv(names, csv_path):
    # Open the CSV file in write mode with utf-8 encoding
    with open(csv_path, 'w', newline='', encoding='utf-8') as csvfile:
        # Create a CSV writer
        writer = csv.writer(csvfile)
        # Write the header
        writer.writerow(["Name"])
        # Write the names
        for name in names:
            writer.writerow([name])

# Call the function to save the names to a CSV file
# Note: Replace 'names.csv' with the desired CSV file name
save_names_to_csv(names, 'names.csv')
```
