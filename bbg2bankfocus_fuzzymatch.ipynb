# **Levenshtein score : Fuzzy matching on multiple files**

```python
import pandas as pd
from fuzzywuzzy import fuzz
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
import nltk

# Define preprocessing function
def preprocess_text(text):
    if isinstance(text, str):
        # Convert text to lowercase
        text = text.lower()
        # Remove punctuation
        text = text.translate(str.maketrans('', '', string.punctuation))
        return text
    else:
        return ''

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Function to tokenize text
def tokenize_text(text):
    tokens = word_tokenize(text)
    return tokens

# Function to remove stopwords
def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens

# Function to stem tokens
def stem_tokens(tokens):
    porter = PorterStemmer()
    stemmed_tokens = [porter.stem(word) for word in tokens]
    return stemmed_tokens

# Function to lemmatize tokens
def lemmatize_tokens(tokens):
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return lemmatized_tokens

# Function to calculate Levenshtein similarity score
def calculate_levenshtein_score(original_text, matching_text):
    levenshtein_score = fuzz.ratio(original_text, matching_text)
    return levenshtein_score

# Read and process each file
file_names = ["Export_BBG lenders1.xlsx", "Export_BBG lenders2.xlsx", "Export_BBG lenders3.xlsx", "Export_BBG lenders4.xlsx", "Export_BBG lenders5.xlsx", "Export_BBG lenders6.xlsx", "Export_BBG lenders7.xlsx", "Export_BBG lenders8.xlsx", "Export_BBG lenders9.xlsx"]

for file_name in file_names:
    print("Processing file:", file_name)
    df = pd.read_excel(file_name)

    # Preprocess text
    df['Normalized Company name'] = df['Company name'].apply(preprocess_text)
    df['Normalized Matched company name'] = df['Matched company name'].apply(preprocess_text)

    # Tokenize text
    df['Tokenized Company name'] = df['Normalized Company name'].apply(tokenize_text)
    df['Tokenized Matched company name'] = df['Normalized Matched company name'].apply(tokenize_text)

    # Remove stopwords
    df['Tokenized Company name'] = df['Tokenized Company name'].apply(remove_stopwords)
    df['Tokenized Matched company name'] = df['Tokenized Matched company name'].apply(remove_stopwords)

    # Stem tokens
    df['Stemmed Company name'] = df['Tokenized Company name'].apply(stem_tokens)
    df['Stemmed Matched company name'] = df['Tokenized Matched company name'].apply(stem_tokens)

    # Lemmatize tokens
    df['Lemmatized Company name'] = df['Tokenized Company name'].apply(lemmatize_tokens)
    df['Lemmatized Matched company name'] = df['Tokenized Matched company name'].apply(lemmatize_tokens)

    # Calculate Levenshtein similarity score
    df['Levenshtein Score'] = df.apply(lambda row: calculate_levenshtein_score(row['Normalized Company name'], row['Normalized Matched company name']), axis=1)

    # Save the processed data to a new CSV file
    output_file_name = file_name.split('.')[0] + '_processed.csv'
    df.to_csv(output_file_name, index=False)
    print("Processed data saved to:", output_file_name)
```

# **Levenshtein score :Optimized for checking matches among all Matched company names when empty**

```python
import pandas as pd
from fuzzywuzzy import fuzz
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

# Function to preprocess text
def preprocess_text(text):
    text = str(text)  # Ensure text is a string
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

# Function to tokenize and remove stopwords
def preprocess_and_tokenize(text):
    text = preprocess_text(text)
    # Tokenize text
    tokens = word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens

# Function to calculate Levenshtein similarity score
def calculate_levenshtein_score(original_text, matching_text):
    levenshtein_score = fuzz.ratio(original_text, matching_text)
    return levenshtein_score

# Read all files and combine into one DataFrame
file_names = ["Export_BBG lenders1.xlsx", "Export_BBG lenders2.xlsx", "Export_BBG lenders3.xlsx", "Export_BBG lenders4.xlsx", "Export_BBG lenders5.xlsx", "Export_BBG lenders6.xlsx", "Export_BBG lenders7.xlsx", "Export_BBG lenders8.xlsx", "Export_BBG lenders9.xlsx"]
dfs = [pd.read_excel(file_name) for file_name in file_names]
combined_df = pd.concat(dfs, ignore_index=True)

# Preprocess text and tokenize
combined_df['Tokenized Company name'] = combined_df['Company name'].apply(preprocess_and_tokenize)
combined_df['Tokenized Matched company name'] = combined_df['Matched company name'].apply(preprocess_and_tokenize)

# Calculate Levenshtein similarity score
combined_df['Levenshtein Score'] = combined_df.apply(lambda row: calculate_levenshtein_score(row['Tokenized Company name'], row['Tokenized Matched company name']), axis=1)

# Function to find potential matches
def find_potential_matches(row):
    company_name = row['Tokenized Company name']
    matched_company_names = combined_df.loc[combined_df.index != row.name, 'Tokenized Matched company name']
    potential_matches = [matched_name for names in matched_company_names for matched_name in names if fuzz.ratio(company_name, matched_name) >= 50]
    return potential_matches

# Apply function to find potential matches
combined_df['Potential Matches'] = combined_df.apply(find_potential_matches, axis=1)

# Save DataFrame to CSV
combined_df.to_csv("combined_processed.csv", index=False)
print("Processed data saved to: combined_processed.csv")
```

# **Levenshtein score : Optimized for multiple cores and batch processing**

```python
# Import necessary libraries
import pandas as pd
from fuzzywuzzy import fuzz
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import string
import os
from multiprocessing import Pool

# Download NLTK resources
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Function to preprocess text
def preprocess_text(text):
    if isinstance(text, str):
        # Convert text to lowercase
        text = text.lower()
        # Remove punctuation
        text = text.translate(str.maketrans('', '', string.punctuation))
        return text
    else:
        return ''

# Function to tokenize text
def tokenize_text(text):
    tokens = word_tokenize(text)
    return tokens

# Function to remove stopwords
def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens

# Function to stem tokens
def stem_tokens(tokens):
    porter = PorterStemmer()
    stemmed_tokens = [porter.stem(word) for word in tokens]
    return stemmed_tokens

# Function to lemmatize tokens
def lemmatize_tokens(tokens):
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return lemmatized_tokens

# Function to calculate Levenshtein similarity score
def calculate_levenshtein_score(original_text, matching_text):
    levenshtein_score = fuzz.ratio(original_text, matching_text)
    return levenshtein_score

# Function to find potential matches for a single company name
def find_potential_matches_single(args):
    company_name, matched_company_names = args
    potential_matches = [matched_name for matched_name in matched_company_names if calculate_levenshtein_score(company_name, matched_name) >= 50]  # Adjust the threshold as needed
    return potential_matches

# Function to find potential matches for all company names
def find_potential_matches(df):
    with Pool() as pool:
        args = [(company, matched) for company, matched in zip(df['Tokenized Company name'], df['Tokenized Matched company name'])]
        potential_matches = pool.map(find_potential_matches_single, args)
        df['Potential Matches'] = potential_matches
        df['Levenshtein Score'] = [max([calculate_levenshtein_score(company, match) for match in matches]) if matches else 0 for company, matches in zip(df['Tokenized Company name'], potential_matches)]
    return df

# Read and process each file
file_names = ["Export_BBG lenders1.xlsx", "Export_BBG lenders2.xlsx", "Export_BBG lenders3.xlsx", "Export_BBG lenders4.xlsx", "Export_BBG lenders5.xlsx", "Export_BBG lenders6.xlsx", "Export_BBG lenders7.xlsx", "Export_BBG lenders8.xlsx", "Export_BBG lenders9.xlsx"]

# Define output directory
output_dir = '/content/drive/My Drive/Colab Notebooks/'

# List to store processed DataFrames
dfs = []

for file_name in file_names:
    print("Processing file:", file_name)
    df = pd.read_excel(file_name)

    # Preprocess text
    df['Normalized Company name'] = df['Company name'].apply(preprocess_text)
    df['Normalized Matched company name'] = df['Matched company name'].apply(preprocess_text)

    # Tokenize text
    df['Tokenized Company name'] = df['Normalized Company name'].apply(tokenize_text)
    df['Tokenized Matched company name'] = df['Normalized Matched company name'].apply(tokenize_text)

    # Remove stopwords
    df['Tokenized Company name'] = df['Tokenized Company name'].apply(remove_stopwords)
    df['Tokenized Matched company name'] = df['Tokenized Matched company name'].apply(remove_stopwords)

    # Stem tokens
    df['Stemmed Company name'] = df['Tokenized Company name'].apply(stem_tokens)
    df['Stemmed Matched company name'] = df['Tokenized Matched company name'].apply(stem_tokens)

    # Lemmatize tokens
    df['Lemmatized Company name'] = df['Tokenized Company name'].apply(lemmatize_tokens)
    df['Lemmatized Matched company name'] = df['Tokenized Matched company name'].apply(lemmatize_tokens)

    # Find potential matches
    df = find_potential_matches(df)

    # Append processed DataFrame to the list
    dfs.append(df)

# Concatenate all processed DataFrames
combined_df = pd.concat(dfs, ignore_index=True)

# Save the processed data to a new CSV file
output_file_name = os.path.join(output_dir, "combined_processed_Levenshtein.csv")
combined_df.to_csv(output_file_name, index=False)
print("Processed data saved to:", output_file_name)
```
